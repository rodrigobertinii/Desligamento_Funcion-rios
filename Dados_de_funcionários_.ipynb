Contagem dos valores distintos das colunas categoricas


```python
df.select_dtypes(include=['object', 'category']).columns

for column in list(df.select_dtypes(include=['object', 'category']).columns):
    print('-'*35)
    print(df[f'{column}'].value_counts())
    print('-'*35)
```

Saber se existe rela√ß√£o entre as colunas


```python
numeric_df = df.select_dtypes(include=[np.number])
correlation_matrix = numeric_df.corr()

plt.figure(figsize=(20, 13))
sns.heatmap(correlation_matrix, square=True,
            annot=True, fmt='.1f', linewidths=0.1, cmap='coolwarm');
```

No caso, n√£o existe nenhuma rela√ß√£o entre as colunas

Analisar os desligamentos por departamentos

```python
fig = px.histogram(df, x='Department', color='Attrition')
fig.update_layout(
    title='Desligamento por Departamento',
    xaxis_title='Departamentos',
    yaxis_title='Frequ√™ncia')
fig.show()
```

Histograma com a distribui√ß√£o das idades do funcion√°rios

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], bins=38, color='green')
plt.title('Distribui√ß√£o de idades')
plt.xlabel('Idade')
plt.ylabel('Frequ√™ncia')
plt.show()
```

Constru√ß√£o de um gr√°fico de pizza sobre a distribui√ß√£o de G√™nero

```python
plt.figure(figsize=(6, 6))
df['Gender'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=['#FF7F7F', '#7FBFFF'])
plt.title('Distribui√ß√£o de G√™nero')
plt.ylabel('')
plt.show()
```

Ver comparativo de quantos funcion√°rios foram desligados e quantos n√£o

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Attrition', data=df, palette={'Yes': '#FF6347', 'No': '#1E90FF'})
plt.title('Contagem de Desligamentos de Funcion√°rios')
plt.xlabel('Desligamentos')
plt.ylabel('Contagem')
plt.show()
```
